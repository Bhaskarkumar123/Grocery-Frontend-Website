```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data-export (1).csv')
print(df)
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df.head()
```

```python
df.columns = df.iloc[0]
df = df.drop(index = 0).reset_index(drop = True)
df.columns = ['Channel group', 'DateHour', 'Users', 'Session' , 'Engaged Seassion', 'Average engagement time per session', 'Engaged sessions per user','Events per session','Engagement rate','Event count']
```

```python
df.head()
```

```python
df.info()
```

```python
df['DateHour'] = pd.to_datetime(df['DateHour'], format ='%Y%m%d%H', errors ='coerce')
numeric_cols = df.columns.drop(['Channel group', 'DateHour'])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors = 'coerce')
df['Hour'] = df['DateHour'].dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

#website seassions and user over time

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10,5))
df.groupby('DateHour')[['Session', 'Users']].sum().plot(ax=plt.gca())
plt.title('Session and USers over time')
plt.xlabel('DateHour')
plt.ylabel('count')
plt.show()
```

#total user by channel

```python

plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Channel group', y='Users', estimator=np.sum, palette='viridis')
plt.title('Total Users busy Channel')
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

#Average engagement time by channel

```python
plt.figure(figsize= (8,5))
sns.barplot(data=df, x='Channel group', y='Average engagement time per session', estimator=np.mean, palette='magma')
plt.title('Avg engagement time by Channel')
plt.xticks(rotation=45)
plt.show()
```

#Engagement RAte Distribution by channel

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Channel group', y='Engaged sessions per user', palette='coolwarm')
plt.title('Engagement rate distribution by channel ')
plt.xticks(rotation=45)
plt.show()
```

```python
session_df = df.groupby('Channel group')[['Session', 'Engaged Seassion']].sum().reset_index()
session_df['Non-Engaged '] = session_df['Session'] - session_df['Engaged Seassion']
session_df_melted = session_df.melt(id_vars='Channel group', value_vars=['Engaged Seassion', 'Non-Engaged'])

plt.figure(figure=(8,5))
sns.borplot(data= session_df_melted, x='Channel group', y='value', hue='variable')
plt.title('Engaged vs Non-Engeged Sessions ')
plt.xticks(rotation=45)
plt.show()
```

#Traffic By hour and channel

```python
heatmap_data = df.groupby(['Hour', 'Channel group'])['Session'].sum().unstack().fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap( heatmap_data, cmap = 'YlGnBu', linewidths=.5 ,annot=True, fmt='0f')
plt.title('Traffic by Hour and Channel')
plt.xlabel('Channel group')
plt.xlabel('Hour of Day')
plt.show()
```
